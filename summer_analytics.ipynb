```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from flask import Flask, request, jsonify
import openai
```

```python
app = Flask(__name__)

openai.api_key = 'sk-proj-gt3Gc4dtPbFXhXgEpdbsT3BlbkFJAUFIjx6RaL5qlQZMX1iT'

def generate_response(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=150,
        n=1,
        stop=None,
        temperature=0.7
    )
    return response.choices[0].text.strip()

@app.route('/generate_response', methods=['POST'])
def generate_response_endpoint():
    data = request.json
    query = data.get('query')
    if not query:
        return jsonify({"error": "No query provided"}), 400

    response = generate_response(query)
    return jsonify({"response": response})

if __name__ == '__main__':
    app.run(debug=True)
```

```python
model_name = "gpt2"

model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

model.eval()
```

```python
def generate_response(prompt, max_length=100):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    with torch.no_grad():
        output = model.generate(input_ids, max_length=max_length, num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(output[0], skip_special_tokens=True)
    return response
```

```python
query = "How can I reset my password?"

response = generate_response(query)

print("Customer Query:", query)
print("AI Response:", response)
```
